# simulation from multiset pCCA model

```python
def sample_sim(num_trial, counts, times=np.zeros((0, 3)),
               weights=None, signals=None):
    num_trial_0 = counts[0].shape[2]
    
    if weights is None:
        weights = [[np.arange(count.shape[1]) for count in counts]
                   for _ in times]

    if signals is None:
        signals = np.ones(times.shape[0])

    pops = [counts[0][:,:,np.random.choice(num_trial_0, num_trial)],
            counts[1][:,:,np.random.choice(num_trial_0, num_trial)]]
    
    for time, weight, signal in zip(times, weights, signals):
        dt = np.random.choice(np.arange(-1,2), num_trial)      
        
        pop_from, pop_to = np.argsort(time[:-1])
        
        for i in np.arange(-1, 2):
            pops[pop_to][np.ix_(np.arange(time[pop_to]+max(i,0), time[pop_to]+time[-1]+max(i,0)),
                         weight[pop_to], dt == i)] += \
            signal * \
            pops[pop_from][np.ix_(np.arange(time[pop_from]-min(i,0), time[pop_from]+time[-1]-min(i,0)),
                         weight[pop_from], dt == i)] 
    
    return pops
```

## check if exact FDR meets the target FDR

```python
alphas = np.array([0.0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30])
```

```python
rejs_sim = np.full((num_sim,) + alphas.shape + (num_time, num_time), False)

console = open('/dev/stdout', 'w')
for iter_sim in range(num_sim):
    console.write('%d-th simulation\n'%(iter_sim+1))
    console.flush()  
    
    prec_sim, corr_sim, precs_perm_sim, corrs_perm_sim =\
    dict(np.load("%s/simulation_result_%.3d.npz"%(temp_dir, iter_sim))).values()
    
    dspr_sim = 2*prec_sim - prec_sim @ corr_sim @ prec_sim
    dsprs_perm_sim = 2*precs_perm_sim \
                     - precs_perm_sim @ corrs_perm_sim @ precs_perm_sim
    std_dsprs_perm_sim = np.std(dsprs_perm_sim, 0)
    pvals_sim = 2 * stats.norm.cdf(-np.abs(dspr_sim / std_dsprs_perm_sim))
    
    _, _, _, rej_sim = \
    inf.STAR_seq_step(pvals_sim[:num_time,num_time:], prop_carve=0.05, alphas=alphas,
                      sigma=1, mux_init=-np.log(0.01))
    rejs_sim[iter_sim] = rej_sim
```

```python
rej_blobs_sim = np.zeros((num_sim,) + alphas.shape + (num_time, num_time))
num_blobs_sim = np.zeros((num_sim,) + alphas.shape)
for iter_sim in range(num_sim):
    for iter_alpha, alpha in enumerate(alphas):
        mask_cp = rejs_sim[iter_sim, iter_alpha].copy()
        iter_blob = 0
        while np.any(mask_cp):
            iter_blob += 1
            queue = [tuple(np.transpose(np.where(mask_cp))[0])]
            mask_cp[queue[0]] = False
            while len(queue) > 0:
                cur = queue.pop(0)
                rej_blobs_sim[iter_sim, iter_alpha, cur[0], cur[1]] = iter_blob
                if cur[0] > 0:
                    if mask_cp[cur[0]-1, cur[1]]:
                        queue.append((cur[0]-1, cur[1]))
                        mask_cp[cur[0]-1, cur[1]] = False
                if cur[0] < num_time-1:
                    if mask_cp[cur[0]+1, cur[1]]:
                        queue.append((cur[0]+1, cur[1]))
                        mask_cp[cur[0]+1, cur[1]] = False
                if cur[1] > 0:
                    if mask_cp[cur[0], cur[1]-1]:
                        queue.append((cur[0], cur[1]-1))
                        mask_cp[cur[0], cur[1]-1] = False
                if cur[1] < num_time-1:
                    if mask_cp[cur[0], cur[1]+1]:
                        queue.append((cur[0], cur[1]+1))
                        mask_cp[cur[0], cur[1]+1] = False
        num_blobs_sim[iter_sim, iter_alpha] = iter_blob
```

```python
fdps_sim = np.sum(rejs_sim > np.sum(H0, 0), (2,3)) \
           / np.maximum(np.sum(rejs_sim, (2,3)), 1)   

fnps_sim = np.sum(rejs_sim < np.sum(H0, 0), (2,3)) \
               / np.maximum(np.sum(H0), 1)

fcds_sim = num_blobs_sim.copy()
for iter_sim in range(num_sim):
    for iter_alpha, _ in enumerate(alphas):
        fcds_sim[iter_sim, iter_alpha] -= np.sum(np.unique(
            rej_blobs_sim[iter_sim, iter_alpha, np.any(H0,0)])> 0)
fcdps_sim = fcds_sim / np.maximum(num_blobs_sim, 1)

fcnps_sim = np.sum(np.all(
    np.logical_or((rejs_sim[:,:,np.newaxis] < H0), np.logical_not(H0))
    , (3,4)), 2) /np.maximum(H0.shape[0], 1)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(alphas, np.mean(fdps_sim, 0), color=(187/256, 0, 0), label=r"$\widehat{FDR}$")
plt.plot(alphas, np.mean(fcdps_sim, 0), color=(102/256, 102/256, 102/256), label = r"$\widehat{FCDR}$")
plt.xlabel(r"Target FDR") ; plt.legend()

plt.subplot(1,2,2)
plt.plot(alphas, np.mean(fnps_sim, 0), color=(187/256, 0, 0), label = r"$\widehat{FNR}$")
plt.plot(alphas, np.mean(fcnps_sim, 0), color=(102/256, 102/256, 102/256), label = r"$\widehat{FCNR}$")
plt.xlabel(r"Target FDR");  plt.legend()
```
